## 4. Gráficos

#### Las librerías más usadas para visualizar datos son seaborn, matplotlib y plotly.


```python
import seaborn as sea
import matplotlib.pyplot as plt
```

### 4.1 Gráficos para una variable 
##### 4.1.1 Variable cuantitativa discreta

```python
# Gráfico de histograma
sea.distplot(datos.Antigüedad, kde=False)

#Títulos para el gráfico y ejes
x=sea.distplot(datos.Antigüedad, kde=False, rug=False) # Atribuimos el nombre "x" al gráfico generado. 

x.set_title("Gráfico de distribución para meses de Antigüedad") # Título
x.set_xlabel('Meses de Antigüedad') # Nombre del eje x
x.set_ylabel('Cantidad de Clientes') # Nombre del eje y
```

##### 4.1.2 Variable cuantitativa continua 

```python
#Gráfico de distribución 
sea.kdeplot(datos.Antigüedad, shade=True)
```

```python
#Gráfico de distribución con histograma
sea.distplot(datos.Antigüedad, hist=True, rug=False)
```

##### 4.1.3 Variable cualitativa nominal (categórica)

```python
sea.catplot(x = "Fuga", y = "Antigüedad", data = datos, kind = "box")
```

El diagrama de caja, también conocido como diagrama de caja y bigote o box plot, resume la distribución de los datos mostrando la posición de sus cuartiles y valores atípicos (outliers). Se puede graficar con seaborn utilizando la función seaborn.catplot y el argumento kind = "box" o con la función seaborn.boxplot. 

Por defecto, tras calcular la mediana y los cuartiles Q1 y Q3, se calcula el rango intercuantil (Q3 - Q1). Los "bigotes" o "whiskers" (la línea que sale de la caja) se extienden hasta cubrir 1.5 veces el rango intercuantil por debajo y por encima de Q1 y de Q3, respectivamente. Todos los puntos fuera de este rango son considerados valores atípicos y son mostrados como puntos individuales.

### 4.3 Usando la librería de matplotlib para graficar funciones

Definimos la siguiente función
$$f(x)=x^2+3$$

```python
# Definimos la función en phyton
def f(x):
    return (x**2)+3
# Creamos el dominio de x
x=np.linspace(0,10)

#Tanto f(x) y x son arreglos.
```

```python
# Los argumentos de la función de plot son arreglos. 
plt.plot(x,f(x))

# Agregando titulos a gráficos y ejes
plt.title('Gráfico para la función $f(x)$')
plt.xlabel('Valor de x')
plt.ylabel('Valor de $f(x)$')
```
